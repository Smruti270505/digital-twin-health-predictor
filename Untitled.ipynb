````python

# DIGITAL TWIN PROJECT: Predicting Human Health Trajectory using Vital Signs


# üéØ Target column (the output we want to predict)
# In a digital twin, this represents the health condition (Healthy / At Risk / Critical)
TARGET = "Risk Category"
FORCE_TASK = "classification"

# üìÇ File paths
DATA_PATH = r"C:\predictiveproject\dataset\human_vital_signs_dataset_2024.csv" # Path to dataset (already uploaded)
SAVE_MODEL_PATH = "./best_model.pkl"
                         # Where to save the final trained model

# üîÅ Random seed for reproducibility
RANDOM_STATE = 42


# Set to False if you want to skip advanced explainability (SHAP) for faster runs
USE_SHAP = True


````

````python
import os
import warnings
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
from IPython.display import display, Markdown

# sklearn essentials
import joblib
from sklearn.model_selection import train_test_split, RandomizedSearchCV, GridSearchCV
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler, OneHotEncoder, OrdinalEncoder
from sklearn.metrics import (mean_squared_error, mean_absolute_error, r2_score,
                             accuracy_score, precision_score, recall_score, f1_score, confusion_matrix)

# Try importing optional ML libraries (safe fallback if not installed)
try:
    import xgboost as xgb
except Exception:
    xgb = None

try:
    import lightgbm as lgb
except Exception:
    lgb = None

# Try import SHAP (explainability) ‚Äî not required, but nice if available
try:
    import shap
except Exception:
    shap = None

# ---------- (C) Notebook & plotting defaults ----------
# Suppress noisy warnings for clean output
warnings.filterwarnings("ignore")

# Matplotlib / Seaborn style
sns.set(style="whitegrid", context="notebook")
plt.rcParams["figure.figsize"] = (10, 6)
plt.rcParams["savefig.dpi"] = 150
pd.set_option("display.max_columns", 120)
pd.set_option("display.max_rows", 200)
pd.set_option("display.precision", 4)

# ---------- (D) Quick check that DATA_PATH and TARGET exist (informational) ----------
print(f"Notebook started at: {datetime.now().isoformat()}")
print(f"DATA_PATH = {DATA_PATH}")
print(f"TARGET = {TARGET}")
print("Optional libs available:", "xgboost" if xgb else "-", "lightgbm" if lgb else "-", "shap" if shap else "-")
````

````python
# üìä Cell 2 ‚Äî Load Dataset & Basic Exploration
# --------------------------------------------

print("üîπ Loading dataset from:", DATA_PATH)

# Load the CSV file into a pandas DataFrame
try:
    df = pd.read_csv(DATA_PATH)
    print("‚úÖ Dataset loaded successfully!")
except FileNotFoundError:
    raise FileNotFoundError(f"‚ùå Could not find file at {DATA_PATH}. Check your path or upload the CSV again.")

# Basic info about dataset
print(f"\nüìè Dataset Shape: {df.shape[0]} rows √ó {df.shape[1]} columns")
display(Markdown("### üßæ First 5 Rows of the Dataset"))
display(df.head())

# Summary info
display(Markdown("### ‚ÑπÔ∏è Dataset Information"))
display(df.info())

# Quick descriptive statistics (numerical and categorical)
display(Markdown("### üìà Descriptive Statistics (Numerical Columns)"))
display(df.describe().T)

display(Markdown("### üî§ Categorical Columns Overview"))
display(df.describe(include=['object', 'category']).T)

# Check missing values
missing_values = df.isnull().sum().sort_values(ascending=False)
missing_df = pd.DataFrame({
    "Missing Values": missing_values,
    "Percentage": (missing_values / len(df) * 100).round(2)
})
display(Markdown("### ‚ö†Ô∏è Missing Values Overview"))
display(missing_df[missing_df["Missing Values"] > 0])

# Check duplicates
duplicates = df.duplicated().sum()
print(f"\nüß¨ Duplicate rows found: {duplicates}")

# Preview column names to confirm TARGET exists
if TARGET not in df.columns:
    print(f"‚ùå TARGET '{TARGET}' not found in dataset columns.")
    print("Available columns are:")
    print(list(df.columns))
else:
    print(f"üéØ Target column '{TARGET}' found! Ready for preprocessing.")
````

````python
# üìà Cell 3 ‚Äî Data Understanding & Target Analysis
# -----------------------------------------------

print(f"üéØ Target column: {TARGET}")

# 1Ô∏è‚É£ Check class balance in Risk Category
display(Markdown("### üß≠ Distribution of Risk Categories"))
if TARGET in df.columns:
    display(df[TARGET].value_counts())
    plt.figure(figsize=(7,5))
    sns.countplot(data=df, x=TARGET, palette="viridis", order=df[TARGET].value_counts().index)
    plt.title("Distribution of Risk Categories")
    plt.xlabel("Risk Category")
    plt.ylabel("Count")
    plt.xticks(rotation=30)
    plt.show()
else:
    raise ValueError(f"Target column '{TARGET}' not found in dataframe.")

# 2Ô∏è‚É£ Separate numeric and categorical columns (excluding the target)
num_cols = df.select_dtypes(include=['int64', 'float64']).columns.tolist()
cat_cols = df.select_dtypes(include=['object', 'category']).columns.tolist()

# Remove unwanted columns (identifiers, timestamps)
exclude_cols = ["Patient ID", "Timestamp", TARGET]
num_cols = [c for c in num_cols if c not in exclude_cols]
cat_cols = [c for c in cat_cols if c not in exclude_cols]

print(f"üî¢ Numeric features ({len(num_cols)}):", num_cols)
print(f"üî§ Categorical features ({len(cat_cols)}):", cat_cols)

# 3Ô∏è‚É£ Correlation matrix for numeric features
display(Markdown("### üîó Correlation Heatmap (Numeric Features)"))
if len(num_cols) > 1:
    plt.figure(figsize=(10,8))
    corr = df[num_cols].corr()
    sns.heatmap(corr, annot=False, cmap="coolwarm", center=0)
    plt.title("Correlation Heatmap of Vital Signs")
    plt.show()
else:
    print("Not enough numeric columns for correlation matrix.")

# 4Ô∏è‚É£ Relationship between a few key vitals and Risk Category
display(Markdown("### üíì Key Vital Signs vs. Risk Category"))
key_features = ["Heart Rate", "Body Temperature", "Oxygen Saturation", "Systolic Blood Pressure", "Diastolic Blood Pressure"]
for col in key_features:
    if col in df.columns:
        plt.figure(figsize=(7,5))
        sns.boxplot(data=df, x=TARGET, y=col, palette="Spectral")
        plt.title(f"{col} vs. Risk Category")
        plt.xticks(rotation=30)
        plt.show()
````

````python
# üßπ Cell 4 ‚Äî Data Preprocessing & Cleaning
# ----------------------------------------

display(Markdown("## üßπ Data Preprocessing & Cleaning"))

# 1Ô∏è‚É£ Drop unneeded columns (IDs, timestamps, etc.)
drop_cols = ["Patient ID", "Timestamp"]
for c in drop_cols:
    if c in df.columns:
        df.drop(columns=[c], inplace=True)
        print(f"üóëÔ∏è Dropped column: {c}")

# 2Ô∏è‚É£ Handle missing values
display(Markdown("### ‚öôÔ∏è Handling Missing Values"))

# Show missing before imputation
missing_before = df.isnull().sum().sum()
print(f"Total missing values before cleaning: {missing_before}")

# Simple strategy: numeric -> median, categorical -> most frequent
for col in df.columns:
    if df[col].dtype in ["float64", "int64"]:
        df[col].fillna(df[col].median(), inplace=True)
    else:
        df[col].fillna(df[col].mode()[0], inplace=True)

missing_after = df.isnull().sum().sum()
print(f"‚úÖ Missing values after cleaning: {missing_after}")

# 3Ô∏è‚É£ Verify no missing values remain
if missing_after == 0:
    print("All missing values handled successfully ‚úÖ")
else:
    print("‚ö†Ô∏è Some missing values remain ‚Äî inspect columns manually if needed.")

# 4Ô∏è‚É£ Encode target variable if it‚Äôs categorical
display(Markdown("### üéØ Encoding Target Variable"))
from sklearn.preprocessing import LabelEncoder
if df[TARGET].dtype == "object" or str(df[TARGET].dtype) == "category":
    le = LabelEncoder()
    df[TARGET] = le.fit_transform(df[TARGET])
    label_map = dict(zip(le.classes_, le.transform(le.classes_)))
    print("‚úÖ Encoded target classes:")
    print(label_map)
else:
    print("Target is numeric ‚Äî no encoding needed.")

# 5Ô∏è‚É£ Split dataset into features (X) and target (y)
X = df.drop(columns=[TARGET])
y = df[TARGET]

print(f"‚úÖ Feature matrix X shape: {X.shape}")
print(f"‚úÖ Target vector y shape: {y.shape}")

# 6Ô∏è‚É£ Identify final numeric and categorical columns for preprocessing pipeline
num_cols = X.select_dtypes(include=['int64', 'float64']).columns.tolist()
cat_cols = X.select_dtypes(include=['object', 'category']).columns.tolist()

print(f"üî¢ Final numeric columns ({len(num_cols)}): {num_cols}")
print(f"üî§ Final categorical columns ({len(cat_cols)}): {cat_cols}")

display(Markdown("### ‚úÖ Data Cleaning Completed"))
````

````python
# üö¶ Cell 5 (Fixed) ‚Äî Train/Test Split & Preprocessing Pipeline
# -------------------------------------------------------------

from sklearn.model_selection import train_test_split
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.ensemble import RandomForestClassifier

# 1Ô∏è‚É£ Train / Test split (stratified for classification)
TEST_SIZE = 0.20
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=TEST_SIZE, random_state=RANDOM_STATE, stratify=y
)

print(f"Train set: X={X_train.shape}, y={y_train.shape}")
print(f"Test  set: X={X_test.shape}, y={y_test.shape}")

# 2Ô∏è‚É£ Define transformers
numeric_transformer = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="median")),
    ("scaler", StandardScaler())
])

# handle both old/new sklearn versions
try:
    categorical_transformer = Pipeline(steps=[
        ("imputer", SimpleImputer(strategy="most_frequent")),
        ("onehot", OneHotEncoder(handle_unknown="ignore", sparse_output=False, drop="if_binary"))
    ])
except TypeError:
    # fallback for older sklearn versions (<1.4)
    categorical_transformer = Pipeline(steps=[
        ("imputer", SimpleImputer(strategy="most_frequent")),
        ("onehot", OneHotEncoder(handle_unknown="ignore", sparse=False, drop="if_binary"))
    ])

# 3Ô∏è‚É£ Build column transformer
preprocessor = ColumnTransformer(transformers=[
    ("num", numeric_transformer, num_cols),
    ("cat", categorical_transformer, cat_cols)
], remainder="drop", verbose_feature_names_out=False)

print("\n‚úÖ Preprocessor created successfully!")
print("Numeric columns:", num_cols)
print("Categorical columns:", cat_cols)

# 4Ô∏è‚É£ Fit-transform
X_train_trans = preprocessor.fit_transform(X_train)
X_test_trans  = preprocessor.transform(X_test)
print(f"Transformed shapes ‚Üí X_train_trans: {X_train_trans.shape}, X_test_trans: {X_test_trans.shape}")

# 5Ô∏è‚É£ Build baseline pipeline (preprocessor + Random Forest)
baseline_pipeline = Pipeline(steps=[
    ("preprocessor", preprocessor),
    ("classifier", RandomForestClassifier(n_estimators=200, random_state=RANDOM_STATE, n_jobs=-1))
])

print("\nüéØ Baseline pipeline ready: [preprocessor ‚ûú RandomForestClassifier]")
````

````python
# üöÄ Cell 6 ‚Äî Train Baseline Model, Evaluate & Save
# -------------------------------------------------

from sklearn.metrics import classification_report, ConfusionMatrixDisplay, roc_curve, auc
import matplotlib.pyplot as plt
import numpy as np
import joblib

# 1) Train
print("Training baseline RandomForest pipeline (this may take a minute)...")
baseline_pipeline.fit(X_train, y_train)
print("‚úÖ Training complete.")

# 2) Predictions
y_pred = baseline_pipeline.predict(X_test)

# Try predict_proba (may be used for ROC AUC)
y_proba = None
try:
    y_proba = baseline_pipeline.predict_proba(X_test)[:, 1]  # probability for positive class
except Exception:
    y_proba = None

# 3) Basic metrics
acc = accuracy_score(y_test, y_pred)
prec = precision_score(y_test, y_pred, average="weighted", zero_division=0)
rec = recall_score(y_test, y_pred, average="weighted", zero_division=0)
f1 = f1_score(y_test, y_pred, average="weighted", zero_division=0)

print("\nüî¢ Basic evaluation metrics (test set):")
print(f"Accuracy : {acc:.4f}")
print(f"Precision: {prec:.4f}")
print(f"Recall   : {rec:.4f}")
print(f"F1 score : {f1:.4f}")

# Detailed classification report
print("\nüìã Classification Report:")
print(classification_report(y_test, y_pred, zero_division=0))

# 4) Confusion matrix
plt.figure(figsize=(6,5))
ConfusionMatrixDisplay.from_predictions(y_test, y_pred, cmap="Blues", normalize=None)
plt.title("Confusion Matrix (counts)")
plt.show()

# 5) ROC curve & AUC (binary only)
if y_proba is not None and len(np.unique(y_test)) == 2:
    fpr, tpr, _ = roc_curve(y_test, y_proba)
    roc_auc = auc(fpr, tpr)
    plt.figure(figsize=(6,5))
    plt.plot(fpr, tpr, lw=2, label=f"AUC = {roc_auc:.3f}")
    plt.plot([0,1],[0,1], linestyle="--", color="grey")
    plt.xlabel("False Positive Rate")
    plt.ylabel("True Positive Rate")
    plt.title("ROC Curve")
    plt.legend(loc="lower right")
    plt.show()
    print(f"ROC AUC: {roc_auc:.4f}")
else:
    print("ROC AUC not computed (need binary classification and predict_proba available).")

# 6) Feature importances (for tree-based classifier)
print("\nüîé Feature importances (top features):")
try:
    # get fitted classifier and preprocessor
    clf = baseline_pipeline.named_steps["classifier"]
    pre = baseline_pipeline.named_steps["preprocessor"]
    importances = clf.feature_importances_
    # Build feature names: numeric names + onehot names for categorical
    feat_names = []
    # numeric
    feat_names.extend(num_cols)
    # categorical one-hot names if available
    if len(cat_cols) > 0:
        try:
            onehot = pre.named_transformers_["cat"].named_steps["onehot"]
            # new sklearn method
            try:
                ohe_names = list(onehot.get_feature_names_out(cat_cols))
            except Exception:
                # fallback older sklearn
                ohe_names = list(onehot.get_feature_names(cat_cols))
            feat_names.extend(ohe_names)
        except Exception:
            # fallback: use cat column names
            feat_names.extend(cat_cols)

    # Trim/sync lengths if needed
    if len(importances) != len(feat_names):
        # As a safe fallback, try using the transformed array shape
        example_trans = pre.transform(X_train.iloc[:1])
        n_feats = example_trans.shape[1]
        # create generic names if mismatch
        feat_names = [f"f_{i}" for i in range(n_feats)]
        importances = importances[:n_feats]

    # Create series and show top 15
    fi_series = pd.Series(importances, index=feat_names).sort_values(ascending=False)
    display(fi_series.head(15))

    # Plot top 12
    top_n = min(12, len(fi_series))
    plt.figure(figsize=(8, top_n*0.4+2))
    sns.barplot(x=fi_series.head(top_n).values, y=fi_series.head(top_n).index)
    plt.title("Top feature importances (Random Forest)")
    plt.xlabel("Importance")
    plt.show()

except Exception as e:
    print("Could not compute feature importances:", e)

# 7) Save trained pipeline
try:
    joblib.dump(baseline_pipeline, SAVE_MODEL_PATH)
    print(f"\nüíæ Saved trained pipeline to: {SAVE_MODEL_PATH}")
except Exception as e:
    print("Could not save model:", e)
````

````python
# üîé Cell 7 ‚Äî Hyperparameter Tuning & Cross-Validation
# ----------------------------------------------------
import time
from sklearn.model_selection import StratifiedKFold, RandomizedSearchCV
from scipy.stats import randint, uniform
from sklearn.ensemble import RandomForestClassifier
import joblib

# ---------- Config ----------
N_ITER = 25            # Number of parameter settings sampled for RandomizedSearch
CV_FOLDS = 4
SCORING = "f1_weighted"
RANDOM_SEARCH_RANDOM_STATE = RANDOM_STATE

# Try to import SMOTE (optional)
use_smote = False
try:
    from imblearn.over_sampling import SMOTE
    from imblearn.pipeline import Pipeline as ImbPipeline
    use_smote = True
    print("imblearn found: SMOTE will be available in CV pipeline.")
except Exception:
    ImbPipeline = None
    SMOTE = None
    print("imblearn not found: SMOTE will be skipped. Install with `pip install imbalanced-learn` if you want it.")

# ---------- Helper function to run randomized search ----------
def run_randomized_search(pipeline, param_distributions, X, y, n_iter=N_ITER, cv=CV_FOLDS, scoring=SCORING):
    skf = StratifiedKFold(n_splits=cv, shuffle=True, random_state=RANDOM_STATE)
    rs = RandomizedSearchCV(pipeline, param_distributions=param_distributions,
                            n_iter=n_iter, scoring=scoring, cv=skf,
                            verbose=2, random_state=RANDOM_SEARCH_RANDOM_STATE, n_jobs=-1, return_train_score=True)
    t0 = time.time()
    rs.fit(X, y)
    t1 = time.time()
    print(f"\nRandomizedSearchCV finished in {(t1-t0)/60:.2f} minutes")
    return rs

# ---------- Define base preprocessor (already built) and models ----------
# We'll tune RandomForest; if xgboost is available, we'll tune it too.
# Use a pipeline that optionally includes SMOTE (only for training folds) when ImbPipeline is available.

# 1) Random Forest param distribution
rf_pipeline = None
if use_smote and ImbPipeline is not None:
    rf_pipeline = ImbPipeline(steps=[
        ("preprocessor", preprocessor),
        ("smote", SMOTE(random_state=RANDOM_STATE)),
        ("clf", RandomForestClassifier(random_state=RANDOM_STATE, n_jobs=-1))
    ])
else:
    rf_pipeline = Pipeline(steps=[
        ("preprocessor", preprocessor),
        ("clf", RandomForestClassifier(random_state=RANDOM_STATE, n_jobs=-1))
    ])

rf_param_dist = {
    "clf__n_estimators": randint(100, 600),
    "clf__max_depth": [None] + list(randint(3, 30).rvs(size=6)),
    "clf__min_samples_split": randint(2, 11),
    "clf__min_samples_leaf": randint(1, 8),
    "clf__class_weight": [None, "balanced", "balanced_subsample"]
}

# 2) XGBoost param distribution (if available)
xgb_rs = None
if xgb is not None:
    if use_smote and ImbPipeline is not None:
        xgb_pipeline = ImbPipeline(steps=[
            ("preprocessor", preprocessor),
            ("smote", SMOTE(random_state=RANDOM_STATE)),
            ("clf", xgb.XGBClassifier(use_label_encoder=False, eval_metric='logloss', random_state=RANDOM_STATE))
        ])
    else:
        xgb_pipeline = Pipeline(steps=[
            ("preprocessor", preprocessor),
            ("clf", xgb.XGBClassifier(use_label_encoder=False, eval_metric='logloss', random_state=RANDOM_STATE))
        ])

    xgb_param_dist = {
        "clf__n_estimators": randint(100, 500),
        "clf__learning_rate": uniform(0.01, 0.3),
        "clf__max_depth": randint(3, 12),
        "clf__subsample": uniform(0.5, 0.5),
        "clf__colsample_bytree": uniform(0.5, 0.5)
    }
else:
    xgb_pipeline = None
    xgb_param_dist = None

# ---------- Run RandomizedSearchCV for RandomForest ----------
print("\n===== Tuning RandomForest =====")
rf_search = run_randomized_search(rf_pipeline, rf_param_dist, X_train, y_train, n_iter=N_ITER, cv=CV_FOLDS)

print("\nBest RF params:")
print(rf_search.best_params_)
print(f"Best RF CV {SCORING}: {rf_search.best_score_:.4f}")

# Evaluate tuned RF on the test set
best_rf = rf_search.best_estimator_
y_pred_rf = best_rf.predict(X_test)
y_proba_rf = None
try:
    y_proba_rf = best_rf.predict_proba(X_test)[:,1]
except Exception:
    y_proba_rf = None

from sklearn.metrics import accuracy_score, f1_score, classification_report
print("\n--- Random Forest test set performance ---")
print("Accuracy :", accuracy_score(y_test, y_pred_rf))
print("F1 (weighted):", f1_score(y_test, y_pred_rf, average="weighted"))
print(classification_report(y_test, y_pred_rf, zero_division=0))

# Save this best RF
os.makedirs("saved_models", exist_ok=True)
rf_path = os.path.join("saved_models", "best_rf_pipeline.pkl")
joblib.dump(best_rf, rf_path)
print(f"Saved tuned RandomForest pipeline to: {os.path.abspath(rf_path)}")

# ---------- Optionally tune XGBoost ----------
if xgb_pipeline is not None:
    print("\n===== Tuning XGBoost =====")
    xgb_search = run_randomized_search(xgb_pipeline, xgb_param_dist, X_train, y_train, n_iter=N_ITER, cv=CV_FOLDS)

    print("\nBest XGB params:")
    print(xgb_search.best_params_)
    print(f"Best XGB CV {SCORING}: {xgb_search.best_score_:.4f}")

    best_xgb = xgb_search.best_estimator_
    y_pred_xgb = best_xgb.predict(X_test)
    print("\n--- XGBoost test set performance ---")
    print("Accuracy :", accuracy_score(y_test, y_pred_xgb))
    print("F1 (weighted):", f1_score(y_test, y_pred_xgb, average="weighted"))
    print(classification_report(y_test, y_pred_xgb, zero_division=0))

    xgb_path = os.path.join("saved_models", "best_xgb_pipeline.pkl")
    joblib.dump(best_xgb, xgb_path)
    print(f"Saved tuned XGBoost pipeline to: {os.path.abspath(xgb_path)}")

# ---------- Compare with baseline (optional) ----------
print("\n===== Baseline vs Tuned comparison (F1 weighted) =====")
baseline_pred = baseline_pipeline.predict(X_test)
print("Baseline F1 (weighted):", f1_score(y_test, baseline_pred, average="weighted"))
print("Tuned RF F1 (weighted):", f1_score(y_test, y_pred_rf, average="weighted"))
if xgb_pipeline is not None:
    print("Tuned XGB F1 (weighted):", f1_score(y_test, y_pred_xgb, average="weighted"))

print("\n‚úÖ Hyperparameter tuning complete. Review saved pipelines in ./saved_models/")
````

````python
# üîÅ Cell 8 ‚Äî Model Comparison, Interpretability & Save Final Model
# ----------------------------------------------------------------
import os
import joblib
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import accuracy_score, f1_score, classification_report, confusion_matrix, roc_curve, auc
from IPython.display import display, Markdown

os.makedirs("saved_models", exist_ok=True)

results = {}

# Helper to evaluate a pipeline or estimator (pipeline expected to include preprocessing)
def evaluate_pipeline(name, model, X_test, y_test):
    try:
        y_pred = model.predict(X_test)
    except Exception as e:
        # If the model expects pre-transformed X (rare), try preprocessor first
        try:
            Xt = model.named_steps["preprocessor"].transform(X_test)
            y_pred = model.named_steps["classifier"].predict(Xt)
        except Exception as e2:
            print(f"Could not get predictions for {name}: {e} / {e2}")
            return None
    res = {}
    res["accuracy"] = accuracy_score(y_test, y_pred)
    res["f1_weighted"] = f1_score(y_test, y_pred, average="weighted", zero_division=0)
    res["classification_report"] = classification_report(y_test, y_pred, zero_division=0)
    res["y_pred"] = y_pred
    # try proba for ROC (binary)
    try:
        proba = model.predict_proba(X_test)
        if proba.shape[1] == 2:
            res["y_proba"] = proba[:,1]
    except Exception:
        res["y_proba"] = None
    return res

# 1) Evaluate baseline pipeline (should exist in memory)
available_models = {}
if 'baseline_pipeline' in globals():
    available_models["Baseline"] = baseline_pipeline

# 2) Try to load tuned models from saved_models (if they exist)
rf_path = os.path.join("saved_models", "best_rf_pipeline.pkl")
xgb_path = os.path.join("saved_models", "best_xgb_pipeline.pkl")
if os.path.exists(rf_path):
    try:
        available_models["Tuned_RF"] = joblib.load(rf_path)
    except Exception as e:
        print("Could not load saved RF:", e)
if os.path.exists(xgb_path):
    try:
        available_models["Tuned_XGBoost"] = joblib.load(xgb_path)
    except Exception as e:
        print("Could not load saved XGB:", e)

# 3) If rf_search exists in memory (from RandomizedSearchCV), pick best_estimator_
if 'rf_search' in globals():
    try:
        available_models["Tuned_RF_from_search"] = rf_search.best_estimator_
    except Exception:
        pass
if 'xgb_search' in globals():
    try:
        available_models["Tuned_XGB_from_search"] = xgb_search.best_estimator_
    except Exception:
        pass

# Ensure at least one model exists
if len(available_models) == 0:
    raise RuntimeError("No models found to evaluate. Ensure baseline_pipeline or tuned models exist.")

# 4) Evaluate each model
print("Evaluating available models on the test set...")
for name, model in available_models.items():
    res = evaluate_pipeline(name, model, X_test, y_test)
    if res is not None:
        results[name] = res
        print(f"\n=== {name} ===")
        print(f"Accuracy: {res['accuracy']:.4f} | F1 (weighted): {res['f1_weighted']:.4f}")
        print(res["classification_report"])

# 5) Comparison chart (F1 weighted)
names = []
f1s = []
for k,v in results.items():
    names.append(k)
    f1s.append(v["f1_weighted"])

plt.figure(figsize=(8,4))
sns.barplot(x=f1s, y=names, palette="magma")
plt.xlabel("F1 (weighted)")
plt.title("Model comparison (F1 weighted)")
plt.xlim(0,1)
plt.show()

# 6) Confusion matrix for best model
best_name = max(results.keys(), key=lambda k: results[k]["f1_weighted"])
best_res = results[best_name]
best_model = available_models[best_name]
print(f"\nSelected best model based on F1: {best_name} (F1={best_res['f1_weighted']:.4f})")

plt.figure(figsize=(6,5))
cm = confusion_matrix(y_test, best_res["y_pred"])
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues")
plt.title(f"Confusion Matrix ‚Äî {best_name}")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()

# 7) ROC & AUC if available (binary)
if best_res.get("y_proba", None) is not None and len(np.unique(y_test)) == 2:
    fpr, tpr, _ = roc_curve(y_test, best_res["y_proba"])
    roc_auc = auc(fpr, tpr)
    plt.figure(figsize=(6,5))
    plt.plot(fpr, tpr, label=f"AUC = {roc_auc:.3f}")
    plt.plot([0,1],[0,1],"--", color="grey")
    plt.xlabel("FPR")
    plt.ylabel("TPR")
    plt.title(f"ROC Curve ‚Äî {best_name}")
    plt.legend(loc="lower right")
    plt.show()
    print(f"ROC AUC: {roc_auc:.4f}")

# 8) Feature importances (if tree-based)
def show_feature_importances(pipeline, top_n=15):
    try:
        clf = None
        # Try to locate classifier inside pipeline
        if hasattr(pipeline, "named_steps") and "clf" in pipeline.named_steps:
            clf = pipeline.named_steps["clf"]
        elif hasattr(pipeline, "named_steps") and "classifier" in pipeline.named_steps:
            clf = pipeline.named_steps["classifier"]
        elif hasattr(pipeline, "steps"):
            # maybe a direct model
            for nm, step in pipeline.steps:
                if nm.lower() in ("classifier", "clf", "model"):
                    clf = step
                    break
        if clf is None:
            print("No classifier object found for importances.")
            return

        importances = getattr(clf, "feature_importances_", None)
        if importances is None:
            print("Model has no feature_importances_ attribute.")
            return

        # get feature names from preprocessor (if pipeline)
        feat_names = []
        if hasattr(pipeline, "named_steps") and "preprocessor" in pipeline.named_steps:
            pre = pipeline.named_steps["preprocessor"]
            # numeric names
            feat_names.extend(num_cols)
            # onehot names
            if len(cat_cols) > 0:
                try:
                    onehot = pre.named_transformers_["cat"].named_steps["onehot"]
                    try:
                        ohe_names = list(onehot.get_feature_names_out(cat_cols))
                    except Exception:
                        ohe_names = list(onehot.get_feature_names(cat_cols))
                    feat_names.extend(ohe_names)
                except Exception:
                    feat_names.extend(cat_cols)
        else:
            # fallback
            feat_names = [f"f{i}" for i in range(len(importances))]

        # Sync sizes
        n = min(len(importances), len(feat_names))
        importances = importances[:n]
        feat_names = feat_names[:n]

        fi = pd.Series(importances, index=feat_names).sort_values(ascending=False)
        display(fi.head(top_n))

        plt.figure(figsize=(8, top_n*0.35+1))
        sns.barplot(x=fi.head(top_n).values, y=fi.head(top_n).index)
        plt.title("Top feature importances")
        plt.xlabel("Importance")
        plt.show()
    except Exception as e:
        print("Error showing feature importances:", e)

print("\nFeature importances for the selected best model (if available):")
show_feature_importances(best_model, top_n=12)

# 9) SHAP explanation (if shap available and tree model)
if 'shap' in globals() and shap is not None:
    try:
        import shap
        # Choose small sample to speed up
        sample_X = X_test.sample(min(500, len(X_test)), random_state=RANDOM_STATE)
        # If pipeline, transform then explain if needed; TreeExplainer can take model and raw data for tree models wrapped in pipeline
        if hasattr(best_model, "named_steps") and "preprocessor" in best_model.named_steps:
            pre = best_model.named_steps["preprocessor"]
            clf = None
            if "classifier" in best_model.named_steps:
                clf = best_model.named_steps["classifier"]
            elif "clf" in best_model.named_steps:
                clf = best_model.named_steps["clf"]
            X_sample_trans = pre.transform(sample_X)
            # Use TreeExplainer if model is tree-based
            if hasattr(clf, "predict_proba") or hasattr(clf, "feature_importances_"):
                explainer = shap.Explainer(clf)  # TreeExplainer if appropriate
                shap_values = explainer(X_sample_trans)
                print("\nSHAP summary (beeswarm):")
                shap.plots.beeswarm(shap_values, max_display=15)
            else:
                print("SHAP: classifier not tree-based or not supported for fast explainer.")
        else:
            # If best_model is a raw estimator
            explainer = shap.Explainer(best_model)
            shap_values = explainer(sample_X)
            shap.plots.beeswarm(shap_values, max_display=15)
    except Exception as e:
        print("SHAP explanation failed or shap not installed:", e)
else:
    print("\nSHAP library not available. Install shap to get explainability plots (pip install shap).")

# 10) Save final model (best by F1)
final_path = os.path.join("saved_models", "final_model.pkl")
try:
    joblib.dump(best_model, final_path)
    print(f"\nüíæ Final selected model saved to: {os.path.abspath(final_path)}")
except Exception as e:
    print("Could not save final model:", e)

display(Markdown("### ‚úÖ Cell 8 completed ‚Äî models compared, best saved as `saved_models/final_model.pkl`"))
````

````python
# üîÅ Cell 12 ‚Äî Overwrite app.py with batch-predict + feature-importance features

app_code = """
import streamlit as st
import pandas as pd
import numpy as np
import joblib
import os
import matplotlib.pyplot as plt
import seaborn as sns
from io import BytesIO

st.set_page_config(page_title="Digital Twin ‚Äî Health Risk Predictor", layout="wide")

# ---- Load model ----
MODEL_PATHS = [
    "saved_models/final_model.pkl",
    "saved_models/best_rf_pipeline.pkl",
    "saved_models/best_xgb_pipeline.pkl",
    "best_model.pkl"
]

def load_model():
    for p in MODEL_PATHS:
        if os.path.exists(p):
            try:
                m = joblib.load(p)
                return m, p
            except Exception as e:
                st.warning(f"Found model at {p} but failed to load: {e}")
    return None, None

model, model_path = load_model()

st.title("ü©∫ Digital Twin ‚Äî Health Risk Predictor (Enhanced)")
st.markdown("Enter vitals manually, upload a CSV for batch predictions, view feature importances, and download results.")

col_main, col_side = st.columns([3,1])

with col_side:
    st.subheader("Model status")
    if model is None:
        st.error("No trained model found. Run the notebook cells to save a model in 'saved_models/'.")
    else:
        st.success(f"Loaded model from: {model_path}")
        st.write(type(model))

# ---- Sidebar manual inputs ----
st.sidebar.header("Manual Patient Vitals")
hr = st.sidebar.slider("Heart Rate (bpm)", 30, 200, 80)
rr = st.sidebar.slider("Respiratory Rate (breaths/min)", 8, 40, 16)
temp = st.sidebar.slider("Body Temperature (¬∞C)", 34.0, 42.0, 36.6, step=0.1)
spo2 = st.sidebar.slider("Oxygen Saturation (%)", 60, 100, 97)
sys_bp = st.sidebar.slider("Systolic BP (mmHg)", 70, 220, 120)
dia_bp = st.sidebar.slider("Diastolic BP (mmHg)", 40, 140, 80)
age = st.sidebar.number_input("Age (years)", 0, 120, 35)
gender = st.sidebar.selectbox("Gender", ["Male", "Female", "Other"])
weight = st.sidebar.number_input("Weight (kg)", 20.0, 200.0, 70.0)
height = st.sidebar.number_input("Height (m)", 1.0, 2.5, 1.7)

derived_hrv = max(0.1, 100.0 / max(1, hr))
pulse_pressure = sys_bp - dia_bp
bmi = round(weight / (height**2), 2)
map_val = (2*dia_bp + sys_bp) / 3.0

input_data = {
    "Heart Rate": hr,
    "Respiratory Rate": rr,
    "Body Temperature": temp,
    "Oxygen Saturation": spo2,
    "Systolic Blood Pressure": sys_bp,
    "Diastolic Blood Pressure": dia_bp,
    "Age": age,
    "Gender": gender,
    "Weight (kg)": weight,
    "Height (m)": height,
    "Derived_HRV": derived_hrv,
    "Derived_Pulse_Pressure": pulse_pressure,
    "Derived_BMI": bmi,
    "Derived_MAP": map_val
}
input_df = pd.DataFrame([input_data])

with col_main:
    st.subheader("Input")
    st.dataframe(input_df)

    # Predict single
    st.markdown("### Prediction (single)")
    if model is not None:
        try:
            pred = model.predict(input_df)[0]
            # guess mapping (update if your mapping differs)
            if isinstance(pred, (np.integer, int)):
                mapping = {0: "High Risk", 1: "Low Risk"}
                st.metric("Predicted Risk", mapping.get(int(pred), str(pred)))
            else:
                st.metric("Predicted Risk", str(pred))
            try:
                probs = model.predict_proba(input_df)[0]
                if len(probs) == 2:
                    st.write(f"High Risk probability: {probs[0]:.3f} ‚Äî Low Risk: {probs[1]:.3f}")
                else:
                    st.write("Class probabilities:", probs)
            except Exception:
                st.info("Probabilities not available for this model.")
        except Exception as e:
            st.error(f"Prediction failed: {e}")
    else:
        st.info("No model loaded.")

    st.markdown('---')

    # ---- Batch predictions ----
    st.subheader("Batch Predictions (CSV upload)")
    st.write("Upload a CSV with the same column names as the dataset. Required columns (example):")
    example_cols = list(input_df.columns)
    st.write(example_cols)

    uploaded = st.file_uploader("Upload CSV for batch prediction", type=["csv"])
    if uploaded is not None:
        try:
            df_upload = pd.read_csv(uploaded)
            st.write("Preview of uploaded data:")
            st.dataframe(df_upload.head())

            # Check for missing expected columns
            missing = [c for c in example_cols if c not in df_upload.columns]
            if len(missing) > 0:
                st.warning(f"Uploaded CSV is missing expected columns: {missing}. Attempting to continue but results may be incorrect.")
            if model is not None:
                # Predict (pipeline should handle preprocessing)
                preds = model.predict(df_upload)
                out = df_upload.copy()
                out["Predicted_Risk"] = preds
                try:
                    probs = model.predict_proba(df_upload)
                    if probs.shape[1] == 2:
                        out["Prob_HighRisk"] = probs[:,0]
                        out["Prob_LowRisk"] = probs[:,1]
                except Exception:
                    pass

                st.write("Sample predictions:")
                st.dataframe(out.head())

                # Download results
                csv_bytes = out.to_csv(index=False).encode("utf-8")
                st.download_button("Download predictions (CSV)", data=csv_bytes, file_name="predictions.csv")
            else:
                st.info("Load a model to run batch predictions.")
        except Exception as e:
            st.error(f"Failed to read or predict on uploaded CSV: {e}")

    st.markdown('---')

    # ---- Feature importances ----
    st.subheader("Feature Importances")
    fi_displayed = False
    if model is not None:
        try:
            clf = None
            if hasattr(model, "named_steps"):
                if "classifier" in model.named_steps:
                    clf = model.named_steps["classifier"]
                elif "clf" in model.named_steps:
                    clf = model.named_steps["clf"]
            elif hasattr(model, "feature_importances_"):
                clf = model

            if clf is not None and hasattr(clf, "feature_importances_"):
                importances = clf.feature_importances_
                feat_names = example_cols.copy()  # hope they align; this matches training notebook
                n = min(len(importances), len(feat_names))
                fi = pd.Series(importances[:n], index=feat_names[:n]).sort_values(ascending=False)
                st.bar_chart(fi.head(10))
                st.write("Top feature importances:")
                st.dataframe(fi.head(15).reset_index().rename(columns={"index": "feature", 0: "importance"}))
                fi_displayed = True
            else:
                st.info("Feature importances not available for this model (non-tree model).")
        except Exception as e:
            st.error(f"Could not compute feature importances: {e}")
    if not fi_displayed:
        st.write("No feature importance plot available.")

    st.markdown('---')

    # ---- Simple trajectory simulator (batch) ----
    st.subheader("Trajectory Simulator (batch preview)")
    steps = st.number_input("Steps", min_value=2, max_value=50, value=8)
    hr_delta = st.number_input("HR change per step", min_value=-10, max_value=20, value=1)
    # create a small batch from input_df across steps
    traj = []
    hr_now = hr
    for i in range(steps):
        r = input_data.copy()
        r["Heart Rate"] = hr_now
        traj.append(r)
        hr_now += hr_delta
    traj_df = pd.DataFrame(traj)
    st.write("Trajectory preview:")
    st.dataframe(traj_df.head())

    if model is not None:
        try:
            preds_traj = model.predict(traj_df)
            mapping = {0: "High Risk", 1: "Low Risk"}
            labels = [mapping.get(int(p), str(p)) for p in preds_traj]
            st.line_chart(pd.DataFrame({
                "Heart Rate": traj_df["Heart Rate"],
                "Risk (encoded)": pd.Series(labels).astype('category').cat.codes
            }))
            st.table(pd.DataFrame({"Step": range(1, len(labels)+1), "Heart Rate": traj_df["Heart Rate"], "Predicted Risk": labels}))
        except Exception as e:
            st.error(f"Trajectory prediction failed: {e}")

with col_side:
    st.markdown("### Help & Notes")
    st.write("- Ensure uploaded CSV uses the same column names as displayed above.")
    st.write("- The model mapping of numeric labels to text (High/Low) assumes 0=High Risk, 1=Low Risk. Adjust mapping if your encoding differs.")
    st.write("- SHAP explanations are not shown in this simplified app. Install `shap` and extend the app if needed.")
    st.write("- Validate with clinical experts before real-world use.")

"""

# Write the updated app.py
with open("app.py", "w", encoding="utf-8") as f:
    f.write(app_code)

print("‚úÖ Updated app.py with batch predictions and feature-importance visualization.")
print("Run: streamlit run app.py")
````
